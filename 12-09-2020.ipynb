## Matplotlib
- It is a cross-platform data visulization library for making 2D plots from data

### Type of Plotting

- Line Plot
- Bar Graph
- Scatter Plot
- Area Plot
- Pie Charts

```python
import matplotlib.pyplot as plt
```

### Lines

```python
plt.plot([1,2,3,4],[2,4,6,8],'r')
plt.title('xy graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Bar Graph
- Used to create a Bar Chart

```python
marks = [89,69,72,67,85]
marks1 = [98,96,77,87,82]
names = ['c','java','python','c++','oracle']
plt.bar(names,marks1,color = ['red','yellow','black','pink','green'],width=0.4)
```

## Scatter plot

```python
plt.scatter([1,2,3,4],[2,4,6,8],color='b')
plt.show()
```

## Area plot

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]

plt.stackplot(days,working,sleeping,colors = ['r','g'],labels = ['Work','Rest'])
plt.legend()
plt.show()
```

## Pie chart

```python
marks = [89,69,72,54,85]
marks1 = [98,96,77,87,82]
names = ['c','java','python','c++','oracle']
plt.pie(marks,labels =names)
plt.legend()
```

```python
rating = [5,3,3.5,4.8]
names = ['Chiranjeevi','Venkatesh','Nagarjun','Balakrishna']
plt.pie(rating,labels=names)
plt.legend()
plt.show()
```
